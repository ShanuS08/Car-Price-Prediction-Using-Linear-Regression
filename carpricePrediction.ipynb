```python
#importing libraries
import pandas as pd
```

```python
#reading csv file
df = pd.read_csv("carprices.csv");
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
#getting dummy variables
data = pd.get_dummies(df['Car Model'])
data
```

```python
# Convert True/False to 1/0
data = data.astype(int)
data
```

```python
#joining two dataframe
merged = pd.concat([df,data],axis='columns')
merged
```

```python
#dropping unwanted columns
final = merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
#model-Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#features
x = final.drop(['Sell Price($)'],axis='columns')
x
```

```python
#label
y = final['Sell Price($)']
y
```

```python
#fitting the model
model.fit(x,y)
```

```python
#calculating score
model.score(x,y)
```

```python
#predicting price of mercedezcbenz
model.predict([[45000,4,0,0]])
```

```python
#predicting price of a BMW X5
model.predict([[86000,7,0,1]])
```
